Pie Chart: A pie chart will be used to display the proportion of participants who felt lonely in the past week versus those who did not, providing a clear visual representation of loneliness prevalence in the sample.

The Chi-Square test will help determine whether there's a significant association between the effort put into connecting with others and feeling lonely. 

We will also calculate Cramér's V to measure the strength of the association between the two variables. A higher Cramér's V value would suggest a stronger association between taking initiative to connect with others and the likelihood of feeling lonely. 

Additionally, we will analyze the relationship across the different levels of CONNECTION_effort_connect to see if there's a pattern in loneliness frequencies, specifically focusing on whether higher connection effort is associated with lower loneliness.

To visualize the relationship between the two variables, we will use a stacked bar chart that shows the proportion of people of felt lonely in the past week for each low/medium/high level of perceived effort to connect with others. This is the most appropriate visualizaion method to compare beteween a categorical and a binary variable.

We assume that the sample size is sufficiently large to ensure the validity of the Chi-Square test, as small sample sizes can lead to inaccurate results or violate the test’s assumptions. We also assume that the responses are reliable and accurately reflect an individual’s effort to connect with others and their experience with loneliness. This includes assuming that respondents' self-reports of their connection efforts and loneliness are not biased or influenced by social desirability or other external factors. Furthermore, when calculating Cramér's V, we are assuming that the relationship between the two categorical variables is not influenced by lurking variables or confounding factors that could distort the association.

**Hypothesis**

Null Hypothesis: There is no association between the level of effort to intentionally connect with others and the frequency of feeling lonely. Rejecting the null hypothesis would suggest that taking initiative to intentionally connect with others is associated with a reduction in loneliness, directly answering the research question by confirming or refuting the idea that taking initiative to intentionally connect with others reduces the frequency of loneliness.

Additionally, the Cramér's V value will tell us the strength of the association between the two variables, with a higher Cramér's V suggesting a stronger association.
